```python
# Import necessary libraries
import splitfolders  # To split the dataset into train, val, and test sets
import tensorflow as tf
from tensorflow.keras.applications import EfficientNetB0, DenseNet121
from tensorflow.keras.applications.efficientnet import preprocess_input as preprocess_input_efficientnet
from tensorflow.keras.applications.densenet import preprocess_input as preprocess_input_densenet
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Split dataset into train, val, and test sets
INPUT_FOLDER = 'chest_xrays'  # Original dataset with subfolders for each class (covid19, pneumonia, etc.)
OUTPUT_FOLDER = 'output_folder_2'  # Where split data will be saved

# Split the dataset into 80% train, 10% val, and 10% test
splitfolders.ratio(INPUT_FOLDER, output=OUTPUT_FOLDER, seed=42, ratio=(.8, .1, .1), group_prefix=None)

# Path to the new split dataset directories
TRAIN_DIR = f'{OUTPUT_FOLDER}/train'
VAL_DIR = f'{OUTPUT_FOLDER}/val'
TEST_DIR = f'{OUTPUT_FOLDER}/test'

# Define image size and batch size
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
NUM_CLASSES = 4  # For COVID-19, Pneumonia, Tuberculosis, and Normal

# Data Preprocessing and Augmentation for EfficientNet
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input_efficientnet,  # EfficientNet specific preprocessing
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input_efficientnet)
test_datagen = ImageDataGenerator(preprocessing_function=preprocess_input_efficientnet)

# Load the data from split directories
train_generator = train_datagen.flow_from_directory(
    TRAIN_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

val_generator = val_datagen.flow_from_directory(
    VAL_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

test_generator = test_datagen.flow_from_directory(
    TEST_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

# Build Transfer Learning Model
def build_model(base_model, num_classes=NUM_CLASSES):
    # Freeze the base model layers
    base_model.trainable = False
    
    # Add custom layers
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dropout(0.5)(x)
    x = Dense(1024, activation='relu')(x)
    x = Dropout(0.5)(x)
    predictions = Dense(num_classes, activation='softmax')(x)
    
    # Create the model
    model = Model(inputs=base_model.input, outputs=predictions)
    
    # Compile the model
    model.compile(optimizer=Adam(learning_rate=0.0001),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model

# EarlyStopping and Learning Rate Scheduler callbacks to prevent overfitting and dynamically reduce the learning rate
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1)

# Training and Evaluation Function
def train_and_evaluate(model, train_generator, val_generator, epochs=20):
    history = model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // BATCH_SIZE,
        validation_data=val_generator,
        validation_steps=val_generator.samples // BATCH_SIZE,
        epochs=epochs,
        callbacks=[early_stopping, lr_scheduler]  # Added EarlyStopping and ReduceLROnPlateau callbacks
    )
    
    # Plot Training History
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Loss')
    plt.legend()

    plt.show()
    
    return history

# Load EfficientNetB0 Model with pre-trained ImageNet weights
base_model_efficientnet = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(*IMG_SIZE, 3))
model_efficientnet = build_model(base_model_efficientnet)

# Training EfficientNetB0
print("Training EfficientNetB0 Model...")
history_efficientnet = train_and_evaluate(model_efficientnet, train_generator, val_generator, epochs=20)

# Unfreeze some layers of EfficientNetB0 and Fine-Tune
for layer in base_model_efficientnet.layers[-50:]:
    layer.trainable = True

# Recompile the model with a lower learning rate for fine-tuning
model_efficientnet.compile(optimizer=Adam(learning_rate=0.00001),
                           loss='categorical_crossentropy',
                           metrics=['accuracy'])

print("Fine-tuning EfficientNetB0 Model...")
history_efficientnet_finetune = train_and_evaluate(model_efficientnet, train_generator, val_generator, epochs=10)

# Evaluate on Test Data
print("Evaluating EfficientNetB0 on test data...")
test_loss, test_acc = model_efficientnet.evaluate(test_generator, verbose=2)
print(f"EfficientNetB0 Test Accuracy: {test_acc:.2f}")

# Confusion Matrix Plotting Function
def plot_confusion_matrix(y_true, y_pred, class_names):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()

# Generate Classification Report and Confusion Matrix for EfficientNetB0
Y_pred_efficientnet = model_efficientnet.predict(test_generator, test_generator.samples // BATCH_SIZE + 1)
y_pred_efficientnet = np.argmax(Y_pred_efficientnet, axis=1)
y_true = test_generator.classes
class_labels = list(test_generator.class_indices.keys())

# Print Classification Report
print("EfficientNetB0 Classification Report:")
print(classification_report(y_true, y_pred_efficientnet, target_names=class_labels))

# Plot Confusion Matrix
plot_confusion_matrix(y_true, y_pred_efficientnet, class_labels)

# DenseNet121 Implementation

# Preprocessing and Data Loading for DenseNet
train_datagen_densenet = ImageDataGenerator(
    preprocessing_function=preprocess_input_densenet,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

val_datagen_densenet = ImageDataGenerator(preprocessing_function=preprocess_input_densenet)
test_datagen_densenet = ImageDataGenerator(preprocessing_function=preprocess_input_densenet)

train_generator_densenet = train_datagen_densenet.flow_from_directory(
    TRAIN_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

val_generator_densenet = val_datagen_densenet.flow_from_directory(
    VAL_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

test_generator_densenet = test_datagen_densenet.flow_from_directory(
    TEST_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

# Load DenseNet121 Model with pre-trained ImageNet weights
base_model_densenet = DenseNet121(weights='imagenet', include_top=False, input_shape=(*IMG_SIZE, 3))
model_densenet = build_model(base_model_densenet)

# Train DenseNet121 Model
print("Training DenseNet121 Model...")
history_densenet = train_and_evaluate(model_densenet, train_generator_densenet, val_generator_densenet, epochs=20)

# Fine-tune the model
for layer in base_model_densenet.layers[-50:]:
    layer.trainable = True

# Recompile the model with a lower learning rate for fine-tuning
model_densenet.compile(optimizer=Adam(learning_rate=0.00001),
                       loss='categorical_crossentropy',
                       metrics=['accuracy'])

print("Fine-tuning DenseNet121 Model...")
history_densenet_finetune = train_and_evaluate(model_densenet, train_generator_densenet, val_generator_densenet, epochs=10)

# Evaluate DenseNet121 Model
print("Evaluating DenseNet121 on test data...")
test_loss_densenet, test_acc_densenet = model_densenet.evaluate(test_generator_densenet, verbose=2)
print(f"DenseNet121 Test Accuracy: {test_acc_densenet:.2f}")

# Generate Classification Report and Confusion Matrix for DenseNet121
Y_pred_densenet = model_densenet.predict(test_generator_densenet, test_generator_densenet.samples // BATCH_SIZE + 1)
y_pred_densenet = np.argmax(Y_pred_densenet, axis=1)
y_true_densenet = test_generator_densenet.classes

# Print Classification Report
print("DenseNet121 Classification Report:")
print(classification_report(y_true_densenet, y_pred_densenet, target_names=class_labels))

# Plot Confusion Matrix
plot_confusion_matrix(y_true_densenet, y_pred_densenet, class_labels)
```
